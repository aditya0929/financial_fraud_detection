```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Select features and target variable
features = df[['comparison_with_avg_expenditure', 'transaction_amount','suspicion_indicator']]
target = df['fraud_indicator']

```

```python


# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create and train the Random Forest model with 10 trees and verbose output
model = RandomForestClassifier(n_estimators=10, random_state=42, verbose=1)
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

# Display results
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_output)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, VotingClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Select features and target variable
features = df[['comparison_with_avg_expenditure', 'transaction_amount']]
target = df['fraud_indicator']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Create base classifiers
rf_model = RandomForestClassifier(random_state=42)
gb_model = GradientBoostingClassifier(random_state=42)
xgb_model = XGBClassifier(random_state=42)

# Create a Voting Classifier with soft voting
voting_classifier = VotingClassifier(
    estimators=[
        ('rf', rf_model),
        ('gb', gb_model),
        ('xgb', xgb_model)
    ],
    voting='soft'  # 'soft' for probability-weighted voting
)

# Train the ensemble model
voting_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = voting_classifier.predict(X_test)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

# Display results
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_output)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Select features and target variable
features = df[['comparison_with_avg_expenditure', 'transaction_amount']]
target = df['fraud_indicator']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Build a more complex neural network
model = Sequential()
model.add(Dense(64, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(16, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X_train_scaled, y_train, epochs=100, batch_size=64, validation_split=0.1, verbose=1)

# Make predictions on the test set
y_pred_probs = model.predict(X_test_scaled)
y_pred = (y_pred_probs > 0.5).astype(int)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

# Display results
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_output)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, Dropout
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Select features and target variable
features = df[['comparison_with_avg_expenditure', 'transaction_amount']]
target = df['fraud_indicator']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Data augmentation: Add small random noise to the features
noise_factor = 0.01
X_train_augmented = X_train_scaled + noise_factor * np.random.normal(size=X_train_scaled.shape)

# Build a more complex neural network
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=X_train.shape[1]))
model.add(Dropout(0.5))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(32, activation='relu'))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the model on the augmented data
model.fit(X_train_augmented, y_train, epochs=100, batch_size=64, validation_split=0.1, verbose=1)

# Make predictions on the test set
y_pred_probs = model.predict(X_test_scaled)
y_pred = (y_pred_probs > 0.5).astype(int)

# Evaluate the model
accuracy = accuracy_score(y_test, y_pred)
classification_report_output = classification_report(y_test, y_pred)

# Display results
print(f"Accuracy: {accuracy}")
print("Classification Report:")
print(classification_report_output)
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM, Conv1D, MaxPooling1D, GlobalAveragePooling1D, Dropout
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Select features and target variable
features = df[['comparison_with_avg_expenditure', 'transaction_amount']]
target = df['fraud_indicator']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(features, target, test_size=0.2, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Reshape data for LSTM (sequence length, number of features)
X_train_lstm = np.reshape(X_train_scaled, (X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_lstm = np.reshape(X_test_scaled, (X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

# Build LSTM model
lstm_model = Sequential()
lstm_model.add(LSTM(64, input_shape=(X_train_lstm.shape[1], X_train_lstm.shape[2])))
lstm_model.add(Dropout(0.5))
lstm_model.add(Dense(1, activation='sigmoid'))
lstm_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train LSTM model
lstm_model.fit(X_train_lstm, y_train, epochs=10, batch_size=64, validation_split=0.1, verbose=1)

# Make predictions on the test set
y_pred_lstm_probs = lstm_model.predict(X_test_lstm)
y_pred_lstm = (y_pred_lstm_probs > 0.5).astype(int)

# Evaluate the LSTM model
accuracy_lstm = accuracy_score(y_test, y_pred_lstm)
classification_report_lstm = classification_report(y_test, y_pred_lstm)

# Display LSTM results
print("LSTM Results:")
print(f"Accuracy: {accuracy_lstm}")
print("Classification Report:")
print(classification_report_lstm)

# Reshape data for CNN (sequence length, number of features, 1)
X_train_cnn = np.reshape(X_train_scaled, (X_train_scaled.shape[0], X_train_scaled.shape[1], 1))
X_test_cnn = np.reshape(X_test_scaled, (X_test_scaled.shape[0], X_test_scaled.shape[1], 1))

# Build CNN model
cnn_model = Sequential()
cnn_model.add(Conv1D(64, kernel_size=3, activation='relu', input_shape=(X_train_cnn.shape[1], X_train_cnn.shape[2])))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Conv1D(32, kernel_size=3, activation='relu'))
cnn_model.add(GlobalAveragePooling1D())
cnn_model.add(Dropout(0.5))
cnn_model.add(Dense(1, activation='sigmoid'))
cnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train CNN model
cnn_model.fit(X_train_cnn, y_train, epochs=100, batch_size=64, validation_split=0.1, verbose=1)

# Make predictions on the test set
y_pred_cnn_probs = cnn_model.predict(X_test_cnn)
y_pred_cnn = (y_pred_cnn_probs > 0.5).astype(int)

# Evaluate the CNN model
accuracy_cnn = accuracy_score(y_test, y_pred_cnn)
classification_report_cnn = classification_report(y_test, y_pred_cnn)

# Display CNN results
print("\nCNN Results:")
print(f"Accuracy: {accuracy_cnn}")
print("Classification Report:")
print(classification_report_cnn)
```

```python
import pandas as pd

# Assuming df is your DataFrame containing the dataset
# Replace 'your_data.csv' with the actual path to your dataset

df = pd.read_csv('/kaggle/input/dataset/TransactionDataset1.csv')

# Assuming 'Transaction Location' is the variable of interest
# Create a new categorical variable based on the conditions
df['Transaction_Location_Category'] = pd.cut(
    df['transaction_location'],
    bins=[-float('inf'), 0.1, 0.2, 1.1],  # Assuming 0 to 0.1 is Unknown, 0.1 to 0.2 is Different, and 0.2 to 1.1 is Similar
    labels=['Unknown', 'Different', 'Similar'],
    include_lowest=True
)

# Calculate fraud rates for each category
fraud_rates = df.groupby('Transaction_Location_Category')['Fraud Indicator'].mean()

# Display the fraud rates
print(fraud_rates)
```
